```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("chinook.db")
```

```python
pd.read_sql_query("select t.trackid, t.name [Track Name], t.genreid, m.mediaTypeId, m.name [Media Type] \
                  from tracks t \
                  join media_types m \
                  on t.mediatypeid=m.mediatypeid", conn)
```

```python
pd.read_sql_query("select t.trackid, t.name [Track Name], t.genreid, m.mediaTypeId, m.name [Media Type] \
                  from tracks t \
                  join media_types m \
                  on t.mediatypeid=m.mediatypeid \
                  where t.genreId=1 \
                  ", conn)
```

```python
pd.read_sql_query("select t.trackid, t.name [Track Name], t.genreid, m.mediaTypeId, m.name [Media Type] \
                  from tracks t \
                  join media_types m \
                  on t.mediatypeid=m.mediatypeid and t.genreId=1 \
                  ", conn)
```

```python
# Whenever you would be GROUP BY clause, 
# There are some rules to be taken care of:

# 1. group by usually comes with a aggregation function (Average, Sum, Median, etc)
# 2. We have two axis/dimensions for the group by clause, X/Left-dimension, Y/Right-dimension
# 3. On X/Left-dimension, you are supposed to used qualitative columns (String column, Categor columns, etc)
# 4. On Y/Right-dimension, you are supposed to use the quantitative columns (Numeric Columns)
```

```python
# Whenever you would be GROUP BY clause, 
# There are some rules to be taken care of:

# 1. group by usually comes with a aggregation function (Average, Sum, Median, Count etc)
# 2. We have two axis/dimensions for the group by clause, X/Left-dimension, Y/Right-dimension
# 3. On X/Left-dimension, you are supposed to used qualitative columns (String column, Categor columns, etc)
# 4. On Y/Right-dimension, you are supposed to use the quantitative columns (Numeric Columns)
```

```python
pd.read_sql_query("select Genreid, Count(*) from tracks \
                    group by genreid", conn)
```

```python
pd.read_sql_query("select t.genreid, g.Name, Count(*) \
                  from tracks t \
                  join genres g \
                  on t.genreid=g.genreid \
                    group by t.genreid", conn)
```

```python
pd.read_sql_query("select t.genreid, g.Name, Count(t.Composer) \
                  from tracks t \
                  join genres g \
                  on t.genreid=g.genreid \
                    group by t.genreid, g.Name", conn)
```

```python
pd.read_sql_query("select t.genreid, t.MediaTypeId, Count(*) \
                  from tracks t \
                  join genres g \
                  on t.genreid=g.genreid \
                  group by t.genreid, t.MediaTypeId", conn)
```

```python
pd.read_sql_query("select t.genreid, g.Name, Avg(t.UnitPrice) \
                  from tracks t \
                  join genres g \
                  on t.genreid=g.genreid \
                    group by t.genreid, g.Name", conn)
```
