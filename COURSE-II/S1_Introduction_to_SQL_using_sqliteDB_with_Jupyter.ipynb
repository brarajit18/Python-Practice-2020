```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("chinook.db")
```

```python
pd.read_sql_query("select * from employees", conn)
```

```python
# select * - show everything
# from "this table"
```

```python
pd.read_sql_query("select * from customers", conn)
```

```python
pd.read_sql_query("select * from customers where Country='Canada'", conn)
```

```python
# SELECT QUERY - SHOW THE DATA
# CREATE QUERY - create the table structures
# UPDATE QUERY - To change the data in tables
# INSERT QUERY - To insert new data into a table
```

```python
# Prominent Data types in SQLITE or othe sql databases

# INTEGER           - NUMERIC DATA
# VARCHAR/NVARCHAR  - STRING/TEXT
```

```python
# integer you can apply some comparison operators, i.e. >, <, =, <=, >=
```

```python
pd.read_sql_query("select * from customers where CustomerId <= 5", conn)
```

```python
pd.read_sql_query("select * from customers where Country='Brazil'", conn)
```

```python
pd.read_sql_query("select * from customers where Country='Brazil' and CustomerId=12", conn)
```

```python
pd.read_sql_query("select * from customers where country='Brazil' and customerid=20", conn)
```

```python
pd.read_sql_query("select * from customers where Country='Brazil' or CustomerId=20", conn)
```

```python
pd.read_sql_query("select customerid, firstname, lastname, country from customers", conn)
```

```python
pd.read_sql_query("select customerid, firstname, lastname, country from customers where country='Brazil'", conn)
```

```python
# Select - show
# Mention culumns name or (*)--all columns
# from tablename
# where -- filtering
```

```python
pd.read_sql_query("select * from customers limit 5", conn) # table may millions of rows
```

```python
pd.read_sql_query("select count(*) from customers where country='Canada'", conn) # table may millions of rows
```

```python
pd.read_sql_query("select count(*) from customers", conn) # table may millions of rows
```

```python
pd.read_sql_query("select * from tracks", conn)
```

```python
pd.read_sql_query("select * from genres", conn)
```

```python
pd.read_sql_query("select * from tracks \
                  join genres \
                  on tracks.genreid=genres.genreid", conn)
```

```python
pd.read_sql_query("select tracks.trackid, tracks.name, tracks.genreid, genres.name from tracks \
                  join genres \
                  on tracks.genreid=genres.genreid", conn)
```

```python
pd.read_sql_query("select t.trackid, t.name, t.genreid, g.name \
                  from tracks t \
                  join genres g \
                  on t.genreid=g.genreid", conn)

# Alias -- Sometimes we use alias to reduce the textsize, sometimes we use alias to mention clear column names
```

```python
pd.read_sql_query("select t.trackid, t.name TrackName, t.genreid, g.name GenreName \
                  from tracks t \
                  join genres g \
                  on t.genreid=g.genreid", conn)
```
