```python
def extractMonth(date_value):
    # This function take the date as input
    # and split the date by delimiting character.
    # In this data, we have two types of date formats.
    # Type A: MM/DD/YYYY
    # Type B: DD-MM-YYYY
    if date_value.find('-') == -1: # Type A
        my_date = date_value.split('/')
        month_value = my_date[0]
    else: # Type B
        my_date = date_value.split('-')
        month_value = my_date[1]
    return month_value
```

```python
months = []
for dd in sales['Order Date']:
    month_value = extractMonth(dd)
    months.append(month_value)
```

```python
# Assume you have a data from 2011 to 2020
# Before 2016, your company was using Software Simpli_Accounting (2011-2015)
# 2016 onwards, your company started using different accounting software (2016-2020)
```

```python
def extractYear(date_value):
    # This function take the date as input
    # and split the date by delimiting character.
    # In this data, we have two types of date formats.
    # Type A: MM/DD/YYYY
    # Type B: DD-MM-YYYY
    if date_value.find('-') == -1: #Type A
        my_date = date_value.split('/')
        year_value = my_date[2]
    else: #Type B
        my_date = date_value.split('-')
        year_value = my_date[2]
    return year_value
```

```python
years = []
for dd in sales['Order Date']:
    years_value = extractYear(dd)
    years.append(years_value)
```
